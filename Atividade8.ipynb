```python
import pandas as pd 
import numpy as np 
```

```python
amostra=np.array([2500,3200,2700,3000,3400,3700,4500,4700,5000,5200,6000,7000])
```

```python
media=np.mean(amostra)
mediana=np.median(amostra)
print(f'Média:{media}')
print(f'Mediana:{mediana}')
```

```python
distancia=(media-mediana)/mediana*100
print(f'dDistância entre a média e a mediana: {distancia:.2f}%')
```

Como eu devo interpretar isso?

Se:

1. Distâncua <= |10%|: a minha média é CONFIÁVEL;

2. |10%| < Distância>|25%|: existe uma uma in influência MODERADA de valores extemos;

3. Distância>=|25%| exixte ALTA influência dos valores extremos

#ATIVIDADE ASSISTIDA: Crie um array com 50numeros aleatórios e repitam a analise sobre a distancia calculada anterior para a nova amostra

```python
amostra= np.random.randint(1, 100, 50)
```

```python
media = np.mean(amostra)
mediana = np.median(amostra)
distancia = (media - mediana) / mediana * 100
print(f"amostra: {amostra}")
print(f"Média: {media}")
print(f"Mediana: {mediana:}")
print(f"Distância entre a média e a mediana: {distancia:.2f}%")
```

```python
#Calculo dos Quartis:
q1=np.percentile(amostra,25)
q2=np.percentile(amostra,50)
q3=np.percentile(amostra,75)

# Exibição:
print(f"Priemiro Quartil(Q1):{q1}")
print(f"Segundo Quartil(Q2):{q2}")
print(f"Terceiro Quartil(Q3):{q3}")
```

```python
#intervalo InterQUARTIL:

if q1 <= media <= q3:
    print("A média está DENTRO do intervalo interquartil, o que sugere confiabilidade.")
else:
    print("A média está FORA do intervalo interquartil, o que sugere MENOS confiabilidade.")
```

```python
#Calculos para verificação do Intervalo InterQuartil (IQR)

iqr=q3-q1
limite_superior=q3+(1.5*iqr)
limite_inferior=q3-(1.5*iqr)

#Resumo Estatistico:
print(f'Média{media}')
print(f'Mediana{mediana}')
print(f'Distância entre média  mediana:{distancia:.2f}%')
print(f"Primeiro Quartil(Q1):{q1}")
print(f"Segundo Quartil(Q2):{q2}")
print(f"Terceiro Quartil(Q3):{q3}")
print(f"Intervalo InterQuartil:{iqr}")
print(f"Limite Superior:{limite_superior}")
print(f"Limite inferior:{limite_inferior}")
```

```python
amostra_df=pd.DataFrame(amostra,columns=['Valores'])
print(amostra_df)
```

```python
#Medidas de Dispersão
valor_maximo=amostra_df['Valores'].max()
valor_minimo=amostra_df['Valores'].min()
amplitude_total=valor_maximo-valor_minimo

print(f'Menor Valor:{valor_minimo}')
print(f'Menor maximo:{valor_maximo}')
print(f'amplitude:{amplitude_total}')
```

```python
import matplotlib.pyplot as plt
#configuração do painel
fig,ax=plt.subplots(1,2,figsize=(12,6))

#Boxplot
ax[0].boxplot(amostra_df[ 'Valores'])
ax[0].set_title('Distribuição de Dados - Boxplot')
ax[0].set_ylabel( 'Valores')

#Exibindo as Medidas
ax[1].text(0.1,0.8,f"Menor Valor:{valor_minimo}")
ax[1].text(0.1,0.7,f"Menor Valor:{valor_maximo}")
ax[1].text(0.1,0.6,f"amplitude:{amplitude_total}")

plt.tight_layout()
plt.show()
```

```python
salarios={'Salários':[2500,2700,3200,4000,4500,4800,5000,5200,6000,8000,15000]}
df_salarios=pd.DataFrame(salarios)
display(df_salarios)
```

```python
amostra= np.random.randint(1, 100, 50)
serie=pd.Series(amostra)
print(serie)
```

```python
#calculando assimetria
assimetria=serie.skew()
print(f'Assimetrica:{assimetria}')
```

```python
#interpretação

if -0.5 <= assimetria <=0.5:
    resultado="Distribuição Simetrica"
elif assimetria >0.5:
    resultado="Distribuição Assimetrica Positiva"
else:
    resultado="Distribuição Assimetrica Negativa"


print(resultado)
    
```

```python
#Calculando a curtose
curtose=df_salarios['Salários'].kurtosis()
print(f'Curtose:{curtose}')
```

```python
if 2.5 <= curtose <=3.5:
    resultado2="Distribuição Mesocúrtica"
elif curtose < 2.5:
    resultado2="Distribuição Platicúrtica"
else:
    resultado2="Distribuição Leptocúrtica"


print(resultado2)
    
```
